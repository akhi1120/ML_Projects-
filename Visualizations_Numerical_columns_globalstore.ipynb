```python
df=pd.read_csv("C:\\Users\\akhil\\Downloads\\Global-Superstore.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# visualizations 

# Line plot

```python
# 1. Line plot
plt.figure(figsize=(50,10))
plt.plot(df["Quantity"],label='Quantity',c='blue')
plt.title("Line plot of Quantity")
plt.xlabel("Index")
plt.ylabel("Quantity")
plt.legend()
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Sales'], df['Discount'], c='green', alpha=1)
plt.title('Scatter Plot of Sales vs Discount')
plt.xlabel('Sales')
plt.ylabel('Discount')
plt.show()
```

# Bar Plot

```python
plt.figure(figsize=(50, 20))
plt.bar(df.index[:50], df['Sales'][:50], color='purple')
plt.title('Bar Plot of Sales (First 50 Rows)')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(20, 10))
plt.hist(df['Sales'],bins=10, color='yellow', alpha=1)
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Index')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.hist(df['Quantity'],bins=10, color='yellow', alpha=1)
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Index')
plt.show()
```

# Box plot

```python
plt.figure(figsize=(10, 5))
plt.boxplot(df['Sales'],labels=['Sales'])
plt.title('Box Plot of Sales')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(df['Profit'],labels=['Profit'])
plt.title('Box Plot of Profit')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(df['Discount'],labels=['Discount'])
plt.title('Box Plot of Discount')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(df['Shipping Cost'],labels=['Shipping Cost'])
plt.title('Box Plot of Shipping Cost')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(df['Quantity'],labels=['Quantity'])
plt.title('Box Plot of Quantity')
plt.ylabel('Value')
plt.show()
```

# Pie chart

```python
market_profit = df.groupby('Market')['Profit'].sum()

# Pie chart for Profit by Market
plt.figure(figsize=(8, 8))
plt.pie(market_profit, labels=market_profit.index, autopct='%1.1f%%')
plt.title('Pie Chart of Profit by Market')
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

# Define the new custom colors for the pie chart
colors = ['#800080','#C71585', '#8B4513']

# Pie chart for Sales by Category with new colors
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Pie Chart of Sales Across Categories')
plt.show()
```
